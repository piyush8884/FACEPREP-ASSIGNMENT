```python

ans=[]
def per(a,b):
    if len(a)==0:
        ans.append(b)
        return
    for i in range(len(a)):
        temp=a[i]
        left=a[0:i]
        right=a[i+1:]
        x=left+right
        per(x,b+temp)
s="abc"
r=' '
per(s,r)
p=[]
for j in ans:
    p.append(j)
print(p)
```
